***Filling the missing values- imputation***


here the values will be filled with the certain numbers

the possible ways are


*   filling the data with the mean or median value if its numerical valu

*   if categorical we will be using mode

*   filing the numerical value with 0 to -999, or some other number that will not occur in the data. this is done so that the machine can reconize the data is not real or is different.

*   filling the categorical value with a new type for the missing values




```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
updated_df=df
updated_df['Age']=updated_df['Age'].fillna(df['Age'].mean())
updated_df.info()
```

```python
df['Gender'].mode()
```

```python
updated_df['Gender']=updated_df['Gender'].fillna(df['Gender'].mode())
updated_df.info()
```
