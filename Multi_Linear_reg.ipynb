```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/50_Startups.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
#divide the data into dependent and independent
#x=df.drop('profit').values
#y=df['profit'].values
#or
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
#feature encoding-OneHot Encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
x= np.array(ct.fit_transform(x))
```

```python
#train and test spilt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

# Assuming x_train and x_test now have one-hot encoded columns at indices 0, 1, 2
# and original numerical features (R&D Spend, Administration, Marketing Spend) at indices 3, 4, 5
# We scale only the numerical features.
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
```

```python
#create LR model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#Prediction
y_pred = lr.predict(x_test)
```

```python
#ploting y_test vs  y_pred
plt.plot(y_test,color='red')
plt.plot(y_pred,color='blue')
plt.title('y_test vs y_pred')
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.show()
```
